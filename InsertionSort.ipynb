```python
import random as rd
def insertion_sort(A):
    print(A)
    for i in range(1,len(A)):
        for j in range(i,0,-1):
            if A[j-1]>A[j]:A[j-1],A[j]=A[j],A[j-1]
            else:break
    print("\n\n")
    return A

mist=[]
for i in range(0,1000):
    x = rd.randint(1,10000)
    mist.append(x)

print(insertion_sort(mist))  
```
